16. **Target_cogs** - Target cost of goods sold.
17. **Target_margin** - Target profit margin.
18. **Target_profit** - Target profit amount.
19. **Target_sales** - Target sales amount.
20. **Total_expenses** - Total expenses incurred.
21. **Type** - Type of coffee (e.g., Decaf, Regular).

```python
def plot_values_by_time(df, column_name):
    df1 = df.copy()
    df1['Date'] = pd.to_datetime(df1['Date'])
    df1.set_index('Date', inplace=True)
    plt.figure(figsize=(14, 8))
    sns.lineplot(data=df1, x=df1.index, y=column_name)
    plt.title(f'{column_name} Over Time')
    plt.xlabel('Date')
    plt.ylabel(column_name)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

def plot_numeric_columns(data, numeric_columns):
    num_columns = len(numeric_columns)
    num_rows = num_columns
    fig, axes = plt.subplots(num_rows, 2, figsize=(15, 5 * num_rows))
    for idx, column in enumerate(numeric_columns):
        row = idx
        sns.violinplot(y=data[column], ax=axes[row, 0])
        axes[row, 0].set_title(f'Violin plot of {column}')
        sns.boxplot(y=data[column], ax=axes[row, 1])
        axes[row, 1].set_title(f'Boxplot of {column}')
    plt.tight_layout()
    plt.show()
    
def get_numeric_columns(data):
    numeric_columns = data.select_dtypes(include=['int', 'float']).columns.tolist()
    return numeric_columns

def plot_category_sum_by_month(df, category_column):
    df = df.copy()
    df.loc[:, 'Date'] = pd.to_datetime(df['Date'])
    
    numeric_cols = df.select_dtypes(include=np.number).columns
    monthly_sum = df.groupby([df['Date'].dt.to_period('M'), category_column])[numeric_cols].sum().reset_index()
    monthly_sum['Date'] = monthly_sum['Date'].dt.to_timestamp()

    pivot_df = monthly_sum.pivot(index='Date', columns=category_column, values='Sales').fillna(0)

    pivot_df = pivot_df.reset_index().melt(id_vars=['Date'], var_name=category_column, value_name='Sales')
    
    plt.figure(figsize=(14, 8))
    sns.barplot(
        data=pivot_df,
        x='Date',
        y='Sales',
        hue=category_column,
        palette='tab10'
    )
    plt.title(f'Monthly Sales Sum by {category_column.capitalize()}')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.legend(title=category_column, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.show()
```

## Marketing and Sales

Is there a direct and significant relationship between every dollar spent on marketing and the resulting sales? Discover how strategic investments in marketing can drive substantial growth in sales and profitability.


```python
df['YearMonth'] = df['Date'].dt.to_period('M')

monthly_sales_marketing = df.groupby('YearMonth').agg({'Sales': 'sum', 'Marketing': 'sum'}).reset_index()

monthly_sales_marketing['YearMonth'] = monthly_sales_marketing['YearMonth'].astype(str)

monthly_sales_marketing_melted = monthly_sales_marketing.melt(id_vars='YearMonth', value_vars=['Sales', 'Marketing'], var_name='Category', value_name='Amount')

plt.figure(figsize=(14, 8))
sns.barplot(data=monthly_sales_marketing_melted, x='YearMonth', y='Amount', hue='Category', palette='viridis')
plt.title('Monthly Sales and Marketing Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()
```

```python
df_before_2014 = df[df['Date'] < '2014-05-01']

monthly_avg_marketing_before_2014 = df_before_2014.resample('M', on='Date')['Marketing'].mean().mean()
monthly_avg_marketing_after_2014 = df_up_to_2014.resample('M', on='Date')['Marketing'].mean().mean()

percentage_change = ((monthly_avg_marketing_after_2014 - monthly_avg_marketing_before_2014) / monthly_avg_marketing_before_2014) * 100

print("Average monthly marketing expenses before May 2014:", monthly_avg_marketing_before_2014)
print("Average monthly marketing expenses after May 2014:", monthly_avg_marketing_after_2014)
print("Percentage change in average monthly marketing expenses:", percentage_change)
```

**A significant change can be seen in the monthly marketing expenses, on a monthly average the cafe chain invested 152% more than it spent on marketing from May 2014.**

```python
correlation = monthly_sales_marketing['Sales'].corr(monthly_sales_marketing['Marketing'])
print(f"Correlation between Sales and Marketing: {correlation:.2f}")

# ×™×¦×™×¨×ª ×’×¨×£ ×¢× ×©× ×™ ×¦×™×¨×™×
fig, ax1 = plt.subplots(figsize=(10, 8))

# ×¦×™×¨ ×¨××©×•×Ÿ - ×ž×›×™×¨×•×ª
sns.lineplot(data=monthly_sales_marketing, x='YearMonth', y='Sales', ax=ax1, label='Sales', color='b', marker='o')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.tick_params(axis='x', rotation=45)

# ×¦×™×¨ ×©× ×™ - ×”×•×¦××•×ª ×¢×œ ×©×™×•×•×§
ax2 = ax1.twinx()
sns.lineplot(data=monthly_sales_marketing, x='YearMonth', y='Marketing', ax=ax2, label='Marketing', color='g', marker='o')
ax2.set_ylabel('Marketing Expenses', color='g')
ax2.tick_params(axis='y', labelcolor='g')

plt.title('Monthly Sales and Marketing Expenses')
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))
plt.show()
```

```python
monthly_data = df.groupby('YearMonth')[['Marketing', 'Sales', 'Profit']].sum()
correlation_matrix_monthly = monthly_data.corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix_monthly, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation between Monthly Marketing, Sales, and Profit')
plt.show()
```

## 2. Popular and Profitable Products and Regions

Here is a list of the most popular and profitable products and regions in the coffee chain, based on sales and profitability data analysis:

- **Popular Products**: Products with the highest total sales.
- **Profitable Products**: Products with the highest gross profit.
- **Popular Regions**: Regions with the highest total sales.
- **Profitable Regions**: Regions with the highest gross profit.

These products and regions indicate strong categories that deserve more investment and focused marketing campaigns to promote them. Additionally, this analysis helps identify areas and products that require improvement to enhance the overall profitability of the company.


```python
product_analysis = df.groupby('Product').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

product_analysis['Sales_Rank'] = product_analysis['Sales'].rank(ascending=False)
product_analysis['Profit_Rank'] = product_analysis['Profit'].rank(ascending=False)

product_analysis['Combined_Rank'] = product_analysis[['Sales_Rank', 'Profit_Rank']].mean(axis=1)

product_analysis_sorted = product_analysis.sort_values(by='Combined_Rank')
product_analysis_melted = product_analysis_sorted.melt(id_vars='Product', value_vars=['Sales', 'Profit'], var_name='Metric', value_name='Amount')

plt.figure(figsize=(14, 8))
sns.barplot(data=product_analysis_melted, x='Product', y='Amount', hue='Metric', palette='viridis')
plt.title('Products by Sales and Profit')
plt.xlabel('Product')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.legend(title='Metric')
plt.show()
```

```python
df['Profit_Percentage'] = (df['Profit'] / df['Sales']) * 100

product_profit_summary = df.groupby('Product').agg(
    Total_Profit=('Profit', 'sum'),
    Total_Sales=('Sales', 'sum')
).reset_index()

product_profit_summary['Profit_Percentage'] = (product_profit_summary['Total_Profit'] / product_profit_summary['Total_Sales']) * 100

total_profit = product_profit_summary['Total_Profit'].sum()
product_profit_summary['Profit_Share'] = (product_profit_summary['Total_Profit'] / total_profit) * 100

product_profit_summary_sorted = product_profit_summary.sort_values(by='Profit_Share', ascending=False)

display(product_profit_summary_sorted)

plt.figure(figsize=(14, 8))
sns.barplot(data=product_profit_summary_sorted, x='Profit_Share', y='Product', errorbar=None, palette='viridis')
plt.title('Profit Share by Product')
plt.xlabel('Profit Share (%)')
plt.ylabel('Product')
plt.grid(True)
plt.show()
```

```python
df['Profit_Percentage'] = (df['Profit'] / df['Sales']) * 100
state_profit_summary = df.groupby('State').agg(
    Total_Profit=('Profit', 'sum'),
    Total_Sales=('Sales', 'sum')
).reset_index()
state_profit_summary['Profit_Percentage'] = (state_profit_summary['Total_Profit'] / state_profit_summary['Total_Sales']) * 100

total_profit = state_profit_summary['Total_Profit'].sum()
state_profit_summary['Profit_Share'] = (state_profit_summary['Total_Profit'] / total_profit) * 100
state_profit_summary_sorted = state_profit_summary.sort_values(by='Profit_Share', ascending=False)
display(state_profit_summary_sorted)

plt.figure(figsize=(14, 8))
sns.barplot(data=state_profit_summary_sorted, x='Profit_Share', y='State', errorbar=None, palette='coolwarm')
plt.title('Profit Share by State')
plt.xlabel('Profit Share (%)')
plt.ylabel('State')
plt.grid(True)
plt.show()
```

## Conclusion 2 - Coffee chain profitability analysis: products and branches by country

**Invest in profitable products:** focus on expanding and marketing particularly profitable products such as **Colombian** and **lemon**.

**Improve the profitability of high-selling, low-margin products:** review the production and marketing costs of **green tea**, **mint** and **amaranth**; Consider pricing adjustments or process improvements.

**Focus on key markets:** Continue to invest in high-performing states like **California** and **Illinois** to increase market presence and sales.
Improve underperforming markets: Reevaluate marketing strategies and adapt products to local preferences in less profitable countries to improve brand awareness and sales.

**Identifying countries with extremely low sales and profit**:
**New Mexico** and **New Hampshire**: lowest sales and profit. The marketing strategy and business operations in these countries should be checked, and if the performance cannot be improved, you should consider closing the business.

#### Summary:
Strategic investment in profitable products and optimization of operations can improve overall profitability and market performance in all branches.

## 3. Targets vs Actuals

---
## <blockquote>  numeric columns </blockquote>


```python
numeric_columns = get_numeric_columns(df)
plot_numeric_columns(df, numeric_columns)
```

The forecasts for sales, product costs, profit, and margins indicate high variability and significant deviations from the planned values, suggesting a need for improvement in the forecasting process.

```python
 df['Cogs_gap'] = df['Cogs'] - df['Target_cogs']
df['Margin_gap'] = df['Margin'] - df['Target_margin']
df['Profit_gap'] = df['Profit'] - df['Target_profit']
df['Sales_gap'] = df['Sales'] - df['Target_sales ']
  
gap_columns = ['Cogs_gap', 'Margin_gap', 'Profit_gap', 'Sales_gap']
plot_numeric_columns(df, gap_columns)
```

```python
gap_summary = df[gap_columns].describe()
gap_summary
```

```python
gap_columns = ['Cogs_gap', 'Margin_gap', 'Profit_gap', 'Sales_gap']
gap_by_state = df.groupby('State')[gap_columns].mean()

gap_by_state_sorted = gap_by_state.sort_values(by='Profit_gap', ascending=False)

gap_by_state_sorted = gap_by_state_sorted.reset_index()
columns_order = ['State', 'Profit_gap', 'Cogs_gap', 'Margin_gap', 'Sales_gap']
gap_by_state_sorted = gap_by_state_sorted[columns_order]

gap_by_state_sorted
```

```python
gap_summary = df.groupby('State')[gap_columns].mean().reset_index()
gap_summary_sorted = gap_summary.sort_values(by='Profit_gap', ascending=False)
gap_summary_melted_sorted = gap_summary_sorted.melt(id_vars='State', var_name='Gap_Type', value_name='Gap_Value')

plt.figure(figsize=(15, 8))
sns.barplot(x='State', y='Gap_Value', hue='Gap_Type', data=gap_summary_melted_sorted)
plt.title('Average Gaps by State (Sorted by Profit Gap)')
plt.xlabel('State')
plt.ylabel('Average Gap Value')
plt.xticks(rotation=90)
plt.legend(title='Gap Type')
plt.tight_layout()
plt.show()
```
